```python
import pandas as pd

df = pd.read_csv("../data/processed/final_dataset.csv")

print("Dataset shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["date"] = pd.to_datetime(df["date"])

df = df.sort_values(["hospital_id", "ward_code", "date"])

df["date_diff"] = df.groupby(["hospital_id", "ward_code"])["date"].diff()

df["date_diff"].value_counts()
```

```python
# Keep only valid daily sequences (1-day difference)
df_clean = df[df["date_diff"] == pd.Timedelta(days=1)].copy()

print("Original shape:", df.shape)
print("Cleaned shape:", df_clean.shape)
```

```python
df_clean["date_diff"].value_counts()
```

```python
# Save cleaned dataset
df_clean.to_csv("../data/processed/final_dataset_clean.csv", index=False)

print("Clean dataset saved successfully")
```
